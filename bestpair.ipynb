```python
import pandas as pd                    # 'usual' package to structure the data in dataframe
import numpy as np   
```

```python
df = pd.read_csv('mergedata.csv')
df['time_tick']=pd.to_datetime(df['time_tick'], format='%d/%m/%Y %H:%M')
df = df.set_index(pd.DatetimeIndex(df['time_tick']))
df
```

```python
corr_mat=df.corr(method ='pearson').apply(lambda x : x.abs())
corr_mat
```

```python
sorted_corr = corr_mat.unstack().sort_values(kind="quicksort", ascending=False)
sc=pd.DataFrame(sorted_corr, columns=["Value"])[64:74]
sc
```
