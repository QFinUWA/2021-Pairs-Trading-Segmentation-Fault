```python

from statsmodels.tsa.stattools import coint
import numpy as np
# Your New Python File

qb = QuantBook()
# QuantBook Analysis Tool
# For more information see https://www.quantconnect.com/docs/research/overview


#tickers to create pairs out of
tickers = ["MMM","AAP","APA","KO"]
for tick in tickers:
    qb.AddEquity(tick)
pairs = []


#create pairs array
for i in range(len(tickers)-1):
    for j in range(i+1,len(tickers)):
        pairs.append([tickers[i],tickers[j],1])

for pair in pairs:    
    print(pair)

```

```python

h1 = qb.History(qb.Securities.Keys, 50, Resolution.Minute)
for pair in pairs:
    score, pval, _ = coint(h1.loc[pair[0]]['close'], h1.loc[pair[1]]['close'])
    pair[2] = pval
    print(pair[0], "and", pair[1], "have a correlation of:", pval)
```

```python
pairs.sort(key=lambda x: x[2])
for pair in pairs:
    print(pair)
```

```python
for pair in pairs:
    if pair[2] > 0.8: #if the correlation is better than given correlation - display that pair
        print(pair)
```
