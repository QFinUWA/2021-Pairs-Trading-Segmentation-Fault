```python
    import csv
    # Finds tickers for given sector
    # NOTE: Make sure to spell the sector right (case-sensitive) - see S&P500.csv for sector spelling
    #SECTORS: ['Industrials', 'Health Care', 'Information Technology', 'Communication Services', 'Consumer Discretionary', 'Utilities',                  'Financials', 'Materials', 'Real Estate', 'Consumer Staples', 'reports', 'Energy']
    def findTickers(sector):
        SP500 = open("S&P500.csv")
        sectorfilename = "SectorTickers/" + sector + ".csv"
        sector_row_list = [["Ticker", "Security", "SEC filings", "GICS Sector", "GICS Sub-Industry"]]
        lines = SP500.readlines()
        SP500.close()
        for line in lines:
            row = line.split(',')[0:5]
            if row[3] == sector:
                sector_row_list.append(row)
        with open(sectorfilename, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(sector_row_list)

    
    #Find tickers for materials sector
    findTickers("reports")
```
