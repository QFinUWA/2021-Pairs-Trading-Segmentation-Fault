```python
    # Finds tickers for given sector
    # NOTE: Make sure to spell the sector right (case-sensitive) - see S&P500.csv for sector spelling
    def findTickers(sector):
        SP500 = open("S&P500.csv")
        lines = SP500.readlines()
        SP500.close()
        tickers = []
        for line in lines:
            if line.split(',')[3] == sector:
                tickers.append(line.split(',')[0])
        print(tickers)
    
    #Find tickers for materials sector
    findTickers("Materials")
```
